```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
unique_values = {}

for col in df.columns:
    unique_values[col] = df[col].value_counts().shape[0]

pd.DataFrame(unique_values, index=['unique value count']).transpose()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
plt.title("correlation of feature")
```

```python
df.drop('Diabetes_binary', axis=1).corrwith(df.Diabetes_binary).plot(kind='bar', grid=True, figsize=(20, 8)
, title="Correlation with Diabetes_binary",color="Orange");
```

```python
df["Diabetes_binary_str"]= df["Diabetes_binary"].replace({0:"NOn-Diabetic",1:"Diabetic"})
```

```python
df
```

```python
df["Diabetes_binary_str"].value_counts()
```

```python

figure1, plot1 = plt.subplots(1,2,figsize=(8,8))

sns.countplot(df['Diabetes_binary_str'],ax=plot1[0])


labels=["non-Diabetic","Diabetic"]

plt.pie(df["Diabetes_binary"].value_counts() , labels =labels ,autopct='%.02f' );
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Smoker", hue="Diabetes_binary", data = df, palette="Set2")
plt.title("Diabetes by Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("Count")
plt.legend(["No Diabetes", "Diabetes"])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Sex", hue="Diabetes_binary", data = df, palette="Set2")
plt.title("Diabetes by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(["No Diabetes", "Diabetes"])
plt.show()
```

```python
age_diabetes = df.groupby("Age")["Diabetes_binary"].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x="Age", y="Diabetes_binary", data=age_diabetes, palette="mako")
plt.title("Diabetes Prevalence by Age Category")
plt.xlabel("Age Category")
plt.ylabel("Proportion of Diabetes Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.boxplot(x="Diabetes_binary", y="BMI", data=df, palette="coolwarm")
plt.title("BMI Distribution by Diabetes Status")
plt.xlabel("Diabetes Status")
plt.ylabel("BMI")
plt.show()
```
